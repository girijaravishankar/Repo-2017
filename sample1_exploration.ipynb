```
fs, data = wavfile.read('sample_1.wav')
```

```
print("Number of data points : {}".format(len(data)))
```

```
def get_snippet(song, song_start, song_end, len_song = 888):
  song_portion_start = round((song_start/len_song)*len(song))
  song_portion_end = round((song_end/len_song)*len(song))
  return song[song_portion_start:song_portion_end]
  
```

```
song_first_five = get_snippet(data.transpose()[0,], 15, 20)
plt.figure(figsize=(25,6))
plt.plot(song_first_five)
plt.show()
```

```
song_first_five = get_snippet(data.transpose()[1,], 15, 20)
plt.figure(figsize=(25,6))
plt.plot(song_first_five)
plt.show()
```

```
sum(data.transpose()[1,] == data.transpose()[0,])
```

```
from scipy import fft
f = fft(data.transpose()[0,])
```

```
f_abs = np.absolute(f)
```

```
plt.figure(figsize=(25,6))
plt.plot([i*(fs/len(f))for i in range(round(len(f)/2))], f[:round(len(f)/2)])
plt.plot([i*(fs/len(f))for i in range(round(len(f)/2))], f_abs[:round(len(f)/2)], 'r')

plt.show()
```

```
plt.figure(figsize=(25,6))
plt.plot([i*(fs/len(f)) for i in range(round(len(f)/2))], 20*np.log10(f_abs[:round(len(f)/2)])-100, 'g')
plt.show()
```
